```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
cars = pd.read_csv("Car details v3.csv")
print("SHAPE:")
print(cars.shape) #returns (number_of_rows, number_of_columns)
print("\nFIRST 5 ROWS:")
print(cars.head())
print("\nDATASET INFO:")
print(cars.info())
print("\nBASIC STATISTICS:")
print(cars.describe())
#mileage, engine, max power, torque, seats needs their dtype to be changed and all of them have missing values 
```

```python
#adjusting the mileage column
cars["mileage"] = cars["mileage"].str.split().str.get(0)
cars["mileage"] = pd.to_numeric(cars["mileage"], errors = "coerce")
#cars.rename(columns = {"mileage_num": "mileage"}, inplace = True)
cars
```

```python
#adjusting the engine column
cars["engine"] = cars["engine"].str.split().str.get(0)
cars["engine"] = pd.to_numeric(cars["engine"], errors="coerce")
cars = cars.rename(columns = {"engine": "engine (CC)"})
cars
```

```python
#adjusting the max_power column
cars["max_power"] = cars["max_power"].str.split().str.get(0)
cars["max_power"] = pd.to_numeric(cars["max_power"], errors="coerce")
cars = cars.rename(columns = {"max_power": "max_power (bhp)"})
cars
```

```python
#A function to return only the first number of the torque column
def extraction(col):
    if pd.isna(col):
        return None
    col = str(col)
    num_str = ""
    for i in col:
        if i.isdigit() or (i == "." and "." not in num_str):
            num_str += i
        elif num_str:
            break
    try:
        return float(num_str)
    except ValueError:
        return None

#second function to find "kgm" rows and then convert the first number to Nm
def kgm_to_nm(col):
    if pd.isna(col):
        return None
    num = extraction(col)
    if num is None:
        return None
    col = str(col).lower()
    if "kgm" in col:
        converted = round(num * 9.80665, 2)
        return converted
    else:
        return num

cars["torque"] = cars["torque"].apply(kgm_to_nm)
cars = cars.rename(columns = {"torque" : "torque (Nm)"})
cars.info()
```

```python
#filling the empty values to make the column "int64"
mode_seats = cars["seats"].mode().iloc[0]
cars["seats"] = cars["seats"].fillna(mode_seats).astype("Int64")
cars
```

```python
cars.to_csv("cleaned_used_cars.csv", index = False)
```
